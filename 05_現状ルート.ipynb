```python
import pandas as pd
import folium
from branca.element import Figure
import time
import csv
```

```python
dfschool = pd.read_csv("evaluation/fukurois_v2.csv", index_col=0)
dfschool
```

```python
dfvehicle = pd.read_csv('evaluation/中部給食センター_車両.csv', index_col = 0)
dfvehicle 
```

```python
#距離
df_distance = pd.read_csv('csv/distance_matrix.csv', header=None)
# Reorder the matrix
distance = df_distance.values.tolist()
distance
```

```python
#移動時間
df_duration = pd.read_csv('csv/duration_matrix.csv', header=None)
# Reorder the matrix
duration = df_duration.values.tolist()
duration
```

```python
# サービス時間
service_time = 5  # 分
```

```python
# Customer count ('0' is depot) 
customer_count = len(dfschool)
# The number of vehicle
vehicle_count = len(dfvehicle)
# The capacity of vehicle
vehicle_capacity = list(dfvehicle['container'])
```

## 食器（洗浄センターから）

```python
track_routes = [[30,3,28],[30,4,5,28],[30,6,28],[30,0,7,28],[30,2,28],[30,1,28],[30,15,16,28],[30,22,26,28],[30,17,18,28],[30,20,25,29],[30,21,29],[30,19,24,23,27,29]]
track_routes
```

```python
track_names = ['中部1','中部2','中部3','中部4','中部5','中部6','中部7','中部8','中部9','調理1','調理2','調理3']
track_names
```

```python
list_colors = ['red', 'blue', 'green', 'purple', 'orange', 'darkred', 'darkblue', 'darkgreen', 'cadetblue', 'pink', 'gray', 'black', 'lightgray']
```

```python
# foliumでサイズ(600, 400)の地図を描画
fig = Figure(width=600, height=400)
mean_lat = dfschool.iloc[:,3].mean()
mean_long = dfschool.iloc[:,4].mean()            
m = folium.Map(location=(mean_lat, mean_long), zoom_start=12)
#m
```

```python
for i in range(len(track_routes)):
    for j in range(len(track_routes[i])):
        if j != 0:
            #print(track_routes[i][j-1], track_routes[i][j])
            with open("root/latlngs_{}-{}.csv".format(dfschool['学校名'][track_routes[i][j-1]], dfschool['学校名'][track_routes[i][j]])) as f:
                reader = csv.reader(f)
                routes = [(float(row[0]), float(row[1])) for row in reader]
            folium.vector_layers.PolyLine(locations=routes, color=list_colors[i]).add_to(m)
            if track_routes[i][j-1] < 28:
                folium.Marker([dfschool.iloc[track_routes[i][j-1],3],dfschool.iloc[track_routes[i][j-1],4]], popup=folium.Popup(folium.IFrame("<p>{}<br/>  {}</p>".format(track_names[i], dfschool.iloc[track_routes[i][j-1],0]),width=150,height=70), max_width=150), icon=folium.Icon(color=list_colors[i])).add_to(m)
folium.Marker([dfschool.iloc[28,3],dfschool.iloc[28,4]], popup=folium.Popup(folium.IFrame("{}".format(dfschool.iloc[28,0]),width=150,height=50), max_width=150)).add_to(m)
folium.Marker([dfschool.iloc[29,3],dfschool.iloc[29,4]], popup=folium.Popup(folium.IFrame("{}".format(dfschool.iloc[29,0]),width=150,height=50), max_width=150)).add_to(m)
folium.Marker([dfschool.iloc[30,3],dfschool.iloc[30,4]], popup=folium.Popup(folium.IFrame("{}".format(dfschool.iloc[30,0]),width=150,height=50), max_width=150)).add_to(m)
m
```

```python
result = []
for i in range(len(track_routes)):
    r = [1, -service_time, 0, 0, 0, 0]
    for j in range(1, len(track_routes[i])):
        r[0] += 1#おとづれる数
        r[1] += duration[track_routes[i][j-1]][track_routes[i][j]]+service_time#移動時間
        r[2] += distance[track_routes[i][j-1]][track_routes[i][j]]#移動距離
        r[3] += abs(dfschool['elevation'][track_routes[i][j-1]] - dfschool['elevation'][track_routes[i][j]])#高度変化
        r[4] += dfschool['demand'][track_routes[i][j]]#積載量
    r[5] = dfvehicle['container'].tolist()[i] if len(dfvehicle) > i else -1 #積載可能量
    result.append(r)
result
```
