### 緯度経度から距離・時間を算出
ルートを検索する日は、

```python
def get_route_info(origin, destination, api_key, departure_time=None):
    # Directions APIのURLを作成
    base_url = "https://maps.googleapis.com/maps/api/directions/json"
    params = {
        "origin": ",".join(map(str, origin)),
        "destination": ",".join(map(str, destination)),
        "key": api_key
    }
    if departure_time:
        params["departure_time"] = departure_time

    # リクエストを送信
    response = requests.get(base_url, params=params)
    data = response.json()
    #print(data)

    # 必要な情報を抽出
    if data['status'] == 'OK':
        route = data['routes'][0]['legs'][0]
        
        # 距離をkmのfloat型で取得
        distance = float(route['distance']['value']) / 1000  # APIはメートル単位で返すため1000で割る
        
        # 時間を分のfloat型で取得
        duration = float(route['duration']['value']) / 60  # APIは秒単位で返すため60で割る
        
        path = route['steps']

        # 緯度経度のリストを取得
        lat_lngs = []
        for step in path:
            start_location = step['start_location']
            end_location = step['end_location']
            lat_lngs.extend([(start_location['lat'], start_location['lng']),
                             (end_location['lat'], end_location['lng'])])

        # 重複を削除
        lat_lngs = list(dict.fromkeys(lat_lngs))

        return distance, duration, lat_lngs
    else:
        return None, None, None
```
